```python
# QUIZ GAME
# Welcome User
print('Welcome to the GK Quiz!')

# Ask User Name
user_name = input('Please say your name: ')

# Greet User
hello_user = 'Hello ' + user_name
print(hello_user)

score = 0

# Question 1
user_ans_1 = input('What is the national animal of India? ')
if user_ans_1.lower() == 'tiger':
    print('Yes! You are correct!')
    score += 10
else:
    print('Sorry! You are incorrect!')
    print('Tiger is the national animal of India')
    score -= 5

# Question 2
user_ans_2 = input('Who wrote the play "Romeo and Juliet"? ')
if user_ans_2.lower() == 'william shakespeare':
    print('Correct! William Shakespeare wrote "Romeo and Juliet"')
    score += 10
else:
    print('Incorrect! The correct answer is William Shakespeare')
    score -= 5

# Question 3
user_ans_3 = input('What is the capital of France? ')
if user_ans_3.lower() == 'paris':
    print('That\'s right! Paris is the capital of France')
    score += 10
else:
    print('Nope! The correct answer is Paris')
    score -= 5

# Question 4
user_ans_4 = input('Which planet is known as the Red Planet? ')
if user_ans_4.lower() == 'mars':
    print('Correct! Mars is the Red Planet')
    score += 10
else:
    print('Incorrect! The Red Planet is Mars')
    score -= 5

# Question 5
user_ans_5 = input('What is the chemical symbol for water? ')
if user_ans_5.lower() == 'h2o':
    print('Yes! H2O is the chemical symbol for water')
    score += 10
else:
    print('Incorrect! The chemical symbol for water is H2O')
    score -= 5

# Question 6
user_ans_6 = input('Who painted the Mona Lisa? ')
if user_ans_6.lower() == 'leonardo da vinci':
    print('Correct! Leonardo da Vinci painted the Mona Lisa')
    score += 10
else:
    print('Nope! The Mona Lisa was painted by Leonardo da Vinci')
    score -= 5

# Question 7
user_ans_7 = input('What is the currency of Japan? ')
if user_ans_7.lower() == 'yen':
    print('That\'s right! Yen is the currency of Japan')
    score += 10
else:
    print('Incorrect! The currency of Japan is yen')
    score -= 5

# Question 8
user_ans_8 = input('Who wrote the novel "To Kill a Mockingbird"? ')
if user_ans_8.lower() == 'harper lee':
    print('Correct! "To Kill a Mockingbird" was written by Harper Lee')
    score += 10
else:
    print('Wrong! The author of "To Kill a Mockingbird" is Harper Lee')
    score -= 5

# Question 9
user_ans_9 = input('What is the largest mammal? ')
if user_ans_9.lower() == 'blue whale':
    print('Yes! The largest mammal is the Blue Whale')
    score += 10
else:
    print('Incorrect! The Blue Whale is the largest mammal')
    score -= 5

# Question 10
user_ans_10 = input('What is the tallest mountain in the world? ')
if user_ans_10.lower() == 'mount everest':
    print('Right! Mount Everest is the tallest mountain in the world')
    score += 10
else:
    print('Wrong! The tallest mountain is Mount Everest')
    score -= 5

# Display Final Score
final_score = 'Your final score is ' + str(score)
print(final_score)
```
