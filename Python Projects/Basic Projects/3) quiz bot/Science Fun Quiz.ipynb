```python
# FUNNY SCIENCE QUIZ GAME
# Welcome Message
print("Welcome to the Funny Science Quiz Game!")

# Ask for Player's Name
player_name = input("What's your name? ")

# Greet the Player
print("Hello, " + player_name + "! Let's have some fun with science!")

# Initialize Score
score = 0

# Question 1
print("Question 1: What is the chemical symbol for water?")
user_ans_1 = input("Your answer: ")
if user_ans_1.lower() == "h2o":
    print("Correct!")
    score += 10
else:
    print("Oops!")

# Question 2
print("Question 2: What is the fastest animal on land?")
user_ans_2 = input("Your answer: ")
if user_ans_2.lower() == "cheetah":
    print("That's right!")
    score += 10
else:
    print("Haha!")

# Question 3
print("Question 3: What is the unit of measurement for energy?")
user_ans_3 = input("Your answer: ")
if user_ans_3.lower() == "joule":
    print("Correct!")
    score += 10
else:
    print("Funny!")

# Question 4
print("Question 4: What is the process of plants making their own food called?")
user_ans_4 = input("Your answer: ")
if user_ans_4.lower() == "photosynthesis":
    print("That's hilarious!")
    score += 10
else:
    print("Good guess!")

# Question 5
print("Question 5: What is the center of an atom called?")
user_ans_5 = input("Your answer: ")
if user_ans_5.lower() == "nucleus":
    print("That's a good one!")
    score += 10
else:
    print("Haha!")

# Question 6
print("Question 6: What is the process of water turning into vapor called?")
user_ans_6 = input("Your answer: ")
if user_ans_6.lower() == "evaporation":
    print("Correct!")
    score += 10
else:
    print("Classic!")

# Question 7
print("Question 7: What is the study of earthquakes called?")
user_ans_7 = input("Your answer: ")
if user_ans_7.lower() == "seismology":
    print("Haha!")
    score += 10
else:
    print("Nice try!")

# Question 8
print("Question 8: What is the outermost layer of the Earth called?")
user_ans_8 = input("Your answer: ")
if user_ans_8.lower() == "crust":
    print("That's clever!")
    score += 10
else:
    print("Funny!")

# Question 9
print("Question 9: What is the chemical symbol for gold?")
user_ans_9 = input("Your answer: ")
if user_ans_9.lower() == "au":
    print("Haha!")
    score += 10
else:
    print("Good one!")

# Question 10
print("Question 10: What is the force that pulls objects toward the center of the Earth?")
user_ans_10 = input("Your answer: ")
if user_ans_10.lower() == "gravity":
    print("Correct!")
    score += 10
else:
    print("That's funny!")

# Display Final Score
print("Thanks for playing, " + player_name + "!")
print("Your final score is: " + str(score) + "/100")
```
