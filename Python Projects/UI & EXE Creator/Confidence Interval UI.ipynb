```python
import warnings
warnings.filterwarnings('ignore')
import tkinter as tk
from tkinter import filedialog as fd

def callback():
    global excel_name
    excel_name = fd.askopenfilename() 
    print(excel_name)

def ci_finder(excel_name, target_ci_cols):
    import pandas as pd
    import numpy as np
    import scipy.stats as st
    
    if '.csv' in excel_name:
        df = pd.read_csv(excel_name)
    else:
        df = pd.read_excel(excel_name)
    
    def ci(data):
        ci_ninty_five = st.norm.interval(alpha=0.95, loc=np.mean(data), scale=st.sem(data))
        ci_ninty_seven = st.norm.interval(alpha=0.97, loc=np.mean(data), scale=st.sem(data))
        ci_ninty_nine = st.norm.interval(alpha=0.99, loc=np.mean(data), scale=st.sem(data))
        print('95%:', ci_ninty_five)
        print('97%:', ci_ninty_seven)
        print('99%:', ci_ninty_nine)
    
    for i in target_ci_cols:
        print(f'Confidence Interval for {i.title()}')
        ci(df[i])
        print('*' * 30)

def main_start(): 
    e_text = entry.get()
    target_ci_cols = e_text.split(',')
    ci_finder(excel_name, target_ci_cols)   

win = tk.Tk()
win.geometry("500x300")
win.title("Confidence Interval Calculator")



# Entry Label
entry_label = tk.Label(win, text="Confidence Interval Calculator",font=('Helvetica',16))
entry_label.place(x=120, y=30)

# Browse Button
browse_btn = tk.Button(win, text='Select Excel File', command=callback)
browse_btn.place(x=200, y=120, width=100, height=30)

# Entry Label
entry_label = tk.Label(win, text="Enter column names (comma-separated):")
entry_label.place(x=50, y=170)

# Entry Field
entry = tk.Entry(win, font=('Helvetica', 12), width=40)
entry.place(x=50, y=200)

# Start Button
start_btn = tk.Button(win, text='Calculate', command=main_start)
start_btn.place(x=200, y=240, width=100, height=30)

win.mainloop()
```
